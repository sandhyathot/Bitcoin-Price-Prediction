```python
df=df.drop(['Date'],axis=1)
```

```python
df.head()
```

```python
df.tail()
```

```python
#prediction for 30days
p_days=30
df['Prediction']=df['Price'].shift(-p_days)
```

```python
df.head()
```

```python
df.tail()
```

```python
#Separate X and Y
x=np.array(df.drop(['Prediction'],axis=1))
x=x[:-p_days]
```

```python
x.shape
```

```python
y=np.array(df['Prediction'])
y=y[:-p_days]
```

```python
y
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
pdays_array=np.array(df.drop(['Prediction'],axis=1))[-p_days:]
```

```python
print(pdays_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=1000,random_state=1)
rf.fit(x_train,y_train)
print('RandomForest Accuracy: {:.2f}%'.format(rf.score(x_test,y_test)*100))
```

```python
#Prediction for test set
y_pred=rf.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
#prediction forecasting for the 30days
rf_predict_30days=rf.predict(pdays_array)
print(rf_predict_30days)
```
